# 6. Manipulating Data

### lower a string

```python
car_sales["Make"].str.lower()
```

```python
car_sales
```

```python
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
```

### In case of missing data

```python
missing_sales = pd.read_csv("missing_data.csv")
missing_sales
```

### you can use fillna() and inplace() tp place a value for the missing values

```python
missing_sales['Odometer (KM)'].fillna(missing_sales['Odometer (KM)'].mean(), inplace= True)
missing_sales
```

```python
missing_sales['Odometer (KM)'].mean()
```

#### dropna() removes the entire row from the DataFrame

```python
missing_sales.dropna(inplace=True)
missing_sales
```

## Creating New Data

### Creating a new Column with a Series

```python
seats_column = pd.Series(['5', '5', '5', '5', '5'])
car_sales['Seats'] = seats_column
car_sales
```

```python
car_sales['Seats'].fillna(5, inplace=True)
car_sales
```

### Creating a new Column with a Python List
#### The list must have as many elements as the index

```python
fuel_economy_list = [5.5, 6, 3.3, 5.8, 9.5, 6, 8, 10, 4.7, 3.6]
car_sales['Fuels Per 100KM']= fuel_economy_list
car_sales
```

### Creating a new Column from an existing column

```python
car_sales['Total Fuel Used'] = car_sales['Odometer (KM)'] / 100 * car_sales['Fuels Per 100KM']
car_sales
```

### Creating a new Column with a single Value

```python
car_sales["No. of Wheels"] = 4
car_sales
```

#### Adding a boolean column in the same way

```python
car_sales["Passed Road Safety"] = True
car_sales
```

```python
car_sales["Passed Road Safety"] = False
car_sales
```

```python
car_sales["Passed Road Safety wrong"] = True
car_sales
```

```python
car_sales["Passed Road Safety"] = True
car_sales
```

### Removing a column

```python
car_sales.drop('Passed Road Safety wrong', axis= 1, inplace= True)
car_sales
```

## Shuffle DataFrame/ Mixup the Database by using Sample

```python
car_sales.sample(frac=1)
```

```python
car_sales_shuffled = car_sales.sample(frac=1)
```

```python
car_sales_shuffled
```

## Resetting the index value in the original order

```python
car_sales_shuffled.reset_index(drop= True, inplace=True)
car_sales_shuffled
```

## Applying a function to a column

```python
car_sales['Odometer (KM)'] = car_sales['Odometer (KM)'].apply(lambda x: x / 1.6)
car_sales
```
