# 2. Matplotlib example Workflow

```python
# 1. Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 3. Setup Plot
fig, ax = plt.subplots(figsize = (5, 5))

# 4. Plot Data
ax.plot(x, y)

# 5. Customize Plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 6. Save and Show
fig.savefig("images/sample_fig.png")
```

## c) Bar Plots

### Making a Plot from a Dictionary

#### Horizontal Bar

```python
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Nut Butter Store",
       xlabel="Items",
       ylabel="Prices");
```

#### Vertical Bar

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

## d) Histograms

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## e) Subplots
### This means Multiple Plots at the same time
### There are two options to make subplots

#### Option-1

```python
fig, ( (ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(50000));
```

#### Option 2

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(50000));
```

## 4.2: matplotlib O O method

```python
over_50 = heart_disease[heart_disease["age"] > 50 ]
over_50.head()
```

```python
len(over_50)
```

### Pyplot Method

```python
over_50.plot(kind = "scatter",
             x = 'age',
             y = 'chol',
             c = 'target',
             figsize = (10, 6));
```

### OO Method (not fully object oriented)

```python
fig, ax = plt.subplots(figsize= (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax)
ax.set_xlim([45,100]);
```

### OO Method(fully object oriented)

```python
fig, ax = plt.subplots(figsize = (10, 6))
# Plot the Data
fig_scatter = ax.scatter(x = over_50["age"],
                         y = over_50["chol"],
                         c = over_50["target"]);
# Customize the Plot
ax.set(title = "Heart Disease",
       xlabel = "Age of the patients",
       ylabel = "Cholesterol Level");
# Add a legend
ax.legend(*fig_scatter.legend_elements(), title="Target");
# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "-.");
```

### Subplot of chol,age, thalach

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10), sharex=True);

# Add data to ax0
fig_scatter0 = ax0.scatter(x = over_50["age"],
                           y = over_50["chol"],
                           c = over_50["target"]);
# Customizing ax0
ax0.set(title="Heart Disease and Cholesterol",
        ylabel = "Cholesterol Level")
# Adding a legend to ax0
ax0.legend(*fig_scatter0.legend_elements(), title = "Target");
# Adding a horizontal line
ax0.axhline(over_50["chol"].mean(),
            linestyle = "-.");

# Add data to ax1
fig_scatter1 = ax1.scatter(x = over_50["age"],
                           y = over_50["thalach"],
                           c = over_50["target"]);
# Customizing ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")
# Adding a legend to ax1
ax1.legend(*fig_scatter1.legend_elements(), title = "Target");
# Adding a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle = "-.");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");
```

# 5. Different styles in matplotlib 

```python
# To see the styles available
plt.style.available
```

## a) Working with a different style
* Changing the dot colors
* Limitting the axis values

```python
plt.style.use('bmh')

# Plotting the subplots


fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10), sharex=True);

# Add data to ax0
fig_scatter0 = ax0.scatter(x = over_50["age"],
                           y = over_50["chol"],
                           c = over_50["target"],
                           cmap = "winter");
# Customizing ax0
ax0.set(title="Heart Disease and Cholesterol",
        ylabel = "Cholesterol Level")

# Limitting the axis values
ax0.set_xlim([50, 80])

# Adding a legend to ax0
ax0.legend(*fig_scatter0.legend_elements(), title = "Target");

# Adding a horizontal line
ax0.axhline(over_50["chol"].mean(),
            linestyle = "-.");

# Add data to ax1
fig_scatter1 = ax1.scatter(x = over_50["age"],
                           y = over_50["thalach"],
                           c = over_50["target"],
                           cmap = "summer");

# Customizing ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# Limitting the axis values
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Adding a legend to ax1
ax1.legend(*fig_scatter1.legend_elements(), title = "Target");

# Adding a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle = "-.");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");
```
